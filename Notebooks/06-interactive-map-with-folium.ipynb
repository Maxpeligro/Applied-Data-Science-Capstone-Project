## Task 1: Mark all launch sites on a map


First, let's try to add each site's location on a map using site's latitude and longitude coordinates


The following dataset with the name `spacex_launch_geo.csv` is an augmented dataset with latitude and longitude added for each site.


```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
```

Now, you can take a look at what are the coordinates for each site.


```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

Above coordinates are just plain numbers that can not give you any intuitive insights about where are those launch sites. If you are very good at geography, you can interpret those numbers directly in your mind. If not, that's fine too. Let's visualize those locations by pinning them on a map.


We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.


```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,


```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
site_map
```

and you should find a small yellow circle near the city of Houston and you can zoom-in to see a larger circle.


Now, let's add a circle for each launch site in data frame `launch_sites`

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
# Assign coordinates to Launch site variables
# Create a dictionary to store launch site → (lat, long) pairs
coords = {
    # For each row in the DataFrame, ignoring the index and keeping only the row data
    row['Launch Site']: (row['Lat'], row['Long'])
    # Loop through every (index, row) pair returned by iterrows()
    for _, row in launch_sites_df.iterrows()
}
# Create a for loop to provide circle and markers for each launch site - everything must be in the for loop
for site in coords:
    # Create circle for each launch site - includes popup to click for site name 
    circle = folium.Circle(coords[site], radius=1000, color='#0000FF', fill=True).add_child(folium.Popup(site))
    # Create markers - Percise pinpoint - discrete launch site location
    marker = folium.map.Marker(coords[site], 
                  icon=DivIcon(icon_size=(20,20),
                               icon_anchor=(0,0), 
                               # Add labels for each marker
                               html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % site, ))
    # Update circle and marker variables with each launch site
    site_map.add_child(circle)
    site_map.add_child(marker)
                  


from IPython.display import HTML
HTML(site_map._repr_html_())
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
# Force-render the Folium map in JupyterLite/Pyodide
from IPython.display import HTML
HTML(site_map._repr_html_())
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# Reuse for loop to easily find whatever Launch Pad coordinates we want
coords = {
    # For each row in the DataFrame, ignoring the index and keeping only the row data
    row['Launch Site']: (row['Lat'], row['Long'])
    # Loop through every (index, row) pair returned by iterrows()
    for _, row in launch_sites_df.iterrows()
}
# Function requires lat1, lon1, so need to break the tuple up and assign variables
lat1,lon1 = coords['CCAFS SLC-40']
lat2,lon2 = 28.56413, -80.56789
distance_coastline = calculate_distance(lat1,lon1, lat2, lon2)
distance_coastline
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example


distance_marker = folium.Marker(
    location=(lat2,lon2),
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
        )
    )

site_map.add_child(distance_marker)
# Force-render the Folium map in JupyterLite/Pyodide
from IPython.display import HTML
HTML(site_map._repr_html_())
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
lat1,lon1 = coords['CCAFS SLC-40']
lat2,lon2 = coords['KSC LC-39A']
lat_rl,long_rl = 28.56416, -80.80255 #nearest railway
lat_hwy,long_hwy = 28.57361, -80.65559 # Nearest Highway - Kenney Parkway North
lat_city,long_city = 28.57291, -80.79813 # Nearest city 

#Calculate distance between CCAFS SLC-40 / KSC LC-39A to nearest railrod, highway, and city
distance_railway_slc_40 = calculate_distance(lat1,lon1,lat_rl,long_rl)
distance_railway_lc_39A = calculate_distance(lat2,lon2,lat_rl,long_rl)
distance_hwy_slc_40 = calculate_distance(lat1,lon1,lat_hwy,long_hwy)
distance_hwy_lc_39A = calculate_distance(lat2,lon2,lat_hwy,long_hwy)
distance_city_slc_40 = calculate_distance(lat1,lon1,lat_city,long_city)
distance_city_lc_39A = calculate_distance(lat2,lon2,lat_city,long_city)

launch_sites = {
    'SLC-40':(lat1, lon1),
    'LC-39A':(lat2, lon2)
}
proximities = {
    'rail':(lat_rl,long_rl),
    'hwy':(lat_hwy,long_hwy),
    'city': (lat_city,long_city)
}
distances = {
    'rail':{
        'SLC-40':distance_railway_slc_40,
        'LC-39A':distance_railway_lc_39A
    },
    'hwy':{
        'SLC-40':distance_hwy_slc_40,
        'LC-39A':distance_hwy_lc_39A
    },
    'city':{
        'SLC-40':distance_city_slc_40,
        'LC-39A':distance_city_lc_39A
    }
}
```

```python
#Add Distance Markers
for prox_name, prox_coord in proximities.items():
    lat_p, long_p = prox_coord
    for site_name, site_coord in launch_sites.items():
        lat_s, long_s = site_coord
        
        distance = distances[prox_name][site_name]
        
        distance_markers = folium.Marker(
            location = (lat_p, long_p),
            icon = DivIcon(
                icon_size = (20,20),
                icon_anchor = (0,0),
                 html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance:.2f} KM</b></div>'
            )
        )
        lines = folium.PolyLine(locations= [(lat_s,long_s),(lat_p,long_p)],
                                weight=2,
                                color='blue'
                               )
        site_map.add_child(distance_markers)
        site_map.add_child(lines)


# Force-render the Folium map in JupyterLite/Pyodide
from IPython.display import HTML
HTML(site_map._repr_html_())
```
